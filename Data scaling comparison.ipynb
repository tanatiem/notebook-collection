```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import scale, minmax_scale, maxabs_scale

import warnings
warnings.filterwarnings('ignore')

plt.style.use('bmh')
```

```python
# data sampling
np.random.seed(69)

data = np.random.randint(-50,50,20)
seq = np.arange(1, len(data) + 1)
```

```python
def display_samples(data):
    plt.figure(figsize=(18,5))
    plt.subplot(131)
    plt.bar(seq, data)
    plt.title('Sample Values')
    plt.subplot(132)
    plt.hist(data, color='coral')
    plt.title('Histrogram')
    plt.subplot(133)
    plt.boxplot(data)
    plt.title('Boxplot');
```

### Data

```python
display_samples(data)
```

### Standardization

```python
data_scaled = scale(data)
display_samples(data_scaled)
```

### Min-Max Scaling

```python
data_minmax = minmax_scale(data)
display_samples(data_minmax)
```

### Max-Abs Scaling

```python
data_maxabs = maxabs_scale(data)
display_samples(data_maxabs)
```
